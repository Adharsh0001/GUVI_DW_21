```python
#-----------------------------------------------------------------------------

# 1
# The block intends to fetch the country to alcohol dataset

def ALCOHOL_DATA():
  import pandas as pd
  df = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/alcohol-consumption/drinks.csv")
  return df

#-----------------------------------------------------------------------------

# 2
# The block intends to fetch the HDI dataset of 2022

def HDI_DATASET():
  import pandas as pd
  HDI = pd.read_csv("/content/csvData.csv")
  return HDI

#-----------------------------------------------------------------------------
# 3
# The block intends to find out the country which serves the alcohol the most

def MOST_ALCOHOL_SERVING_COUNTRY(df):

  # 3.1 - Creates list of each column
  Country,Beer,Spirit,Wine,Pure_Alcohol = Creating_list_of_each_elements(df)
  K = ["Beer","Spirit","Wine","Pure_Alcohol"]

  # 3.2 - Gets the shape of alcohol dataframe
  A = Shape(df)

  # 3.3 - Calculates the totla alcohol served by each country
  Total = Total_Alcohol_consumption(A,Beer,Spirit,Wine,Pure_Alcohol)

  #3.4 - Gets which country serves most of the alcohol
  masc = Most_Alcohol_Serving_Country (Country,Total)
  return Total,Country,A,Beer,Spirit,Wine,Pure_Alcohol,K
#-----------------------------------------------------------------------------

# 3.1
# The block intends to create list of each columns

def Creating_list_of_each_elements(df):
  Country =df.iloc[:,0]
  Beer = df.iloc[:,1]
  Spirit = df.iloc[:,2]
  Wine = df.iloc[:,3]
  Pure_Alcohol = df.iloc[:,4]
  return Country,Beer,Spirit,Wine,Pure_Alcohol

#-----------------------------------------------------------------------------

# 3.2
# The block intends to calculate the row * column dimension

def Shape(df):
  A = df.shape
  return A

#-----------------------------------------------------------------------------

# 3.3 
# The block intends to calculate country wise total alcohol serving

def Total_Alcohol_consumption(A,Beer,Spirit,Wine,Pure_Alcohol):
  L = []
  Z = int(A[0])
  for i in range(Z):
    B = int(Beer[i])
    C = int(Spirit[i])
    D = int(Wine[i])
    E = int(Pure_Alcohol[i])
    Total = B+C+D+E
    L.append(Total)
  return L

#-----------------------------------------------------------------------------

# 3.4
# The block intends to find out which country serves most of the alcohol

def Most_Alcohol_Serving_Country (Country,Total):
  import numpy as np
  masc = np.max(Total)
  masc_index = Total.index(masc)
  print("The major alcohol serving country is" ,Country[masc_index])

#-----------------------------------------------------------------------------

# 4
# The block intends to find out which alcohol has been served more
  
def Alcohol_consumed_most(A,Beer,Spirit,Wine,Pure_Alcohol,K):
  import numpy as np
  Z = int(A[0])
  U= []
  N = 0
  P = 0
  R = 0
  T = 0
  for i in range(Z):
    M = int(Beer[i])
    O = int(Spirit[i])
    Q = int(Wine[i])
    S = int(Pure_Alcohol[i])
    N = M+N
    P = O+P
    R = Q+R
    T = S+T
  U.append(N)
  U.append(P)
  U.append(R)
  U.append(T)
  Most_wanted_Alcohol = np.max(U)
  MWA_index = U.index(Most_wanted_Alcohol)
  print("The most consumed alcohol is ", K[MWA_index])

#-----------------------------------------------------------------------------

# 5
# The block intends to create an relationship between HDI and alcohol dataset

def Relationship_Alcohol_HDI(HDI,Total,Country):

  #5.1 - Get the required HDI dataset
  HDI_TIER = HDI_DATA_TIER(HDI)

  #5.2 - Get the required Alcohol dataset
  Alcohol = Alcohol_DS(Total,Country)

  #5.3 - Merging the HDI and Alcohol dataset
  Re = Merging_Alcohol_HDI(Alcohol,HDI_TIER)
  return Re

#-----------------------------------------------------------------------------

# 5.1
# The block intends get the required HDI dataset

def HDI_DATA_TIER(HDI):
  import pandas as pd
  HDI_TIER_C= HDI.iloc[:,0]
  HDI_TIER_S = HDI.iloc[:,2]
  HDI_TIER_C_DF = pd.DataFrame(HDI_TIER_C)
  HDI_TIER_S_DF = pd.DataFrame(HDI_TIER_S)
  HDI_TIER = HDI_TIER_C_DF.join(HDI_TIER_S_DF)
  return HDI_TIER

#-----------------------------------------------------------------------------

# 5.2
# The block intends get the required Alcohol dataset

def Alcohol_DS(Total,Country):
  import pandas as pd
  Total_consumption = pd.DataFrame(Total)
  Country_df = pd.DataFrame(Country)
  Alcohol = Country_df.join(Total_consumption)
  return Alcohol

#-----------------------------------------------------------------------------

# 5.3
# The block intends merge and give an relationship of HDI and Alcohol dataset

def Merging_Alcohol_HDI(Alcohol,HDI_TIER):
  import pandas as pd
  Re = pd.merge(Alcohol,HDI_TIER,on = "country",how ="inner")
  RE = Re.sort_values(by=[0],ascending =False)
  return RE


#-----------------------------------------------------------------------------

# The main function is the start of the program

def main():
  #1 - Getting the country and alcohol dataset
  df = ALCOHOL_DATA()

  #2 - Getting the HDI dataset
  HDI= HDI_DATASET()

  #3 - Country which serves most of the alcohol
  Total,Country,A,Beer,Spirit,Wine,Pure_Alcohol,K = MOST_ALCOHOL_SERVING_COUNTRY(df)

  #4 - Alcohol which is served more
  Alcohol_consumed_most(A,Beer,Spirit,Wine,Pure_Alcohol,K)

  #5 - Getting the relationship between HDI and Alcohol dataset
  Rel = Relationship_Alcohol_HDI(HDI,Total,Country)
  print(Rel)

#-----------------------------------------------------------------------------

# The program starts witht this main block and executes the function related to it.
main()

#-----------------------------------------------------------------------------
```
